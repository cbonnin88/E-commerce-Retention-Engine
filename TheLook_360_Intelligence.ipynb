```python
%%writefile app.py
import streamlit as st
import pandas as pd
import plotly.express as px
from google.cloud import bigquery
from google.colab import auth

# Page Configuratioin using HEX codes for a personal touch
st.set_page_config(page_title='TheLook: Customer 360',layout='wide')

st.markdown(f"""
 <style>
    .main {{ background-color: #1B1F3B; color: #FFFFFF; }}
    .stMetric {{ background-color: #2D3748; padding: 15px; border-radius: 10px; border-left: 5px solid #27C7C1; }}
    </style>
""",unsafe_allow_html=True)

st.title('üéØ TheLook: Customer Intelligence Tool')

# Sidebar for User Input
st.sidebar.header('Search Parameters')
user_id = st.sidebar.number_input('Enter User ID',min_value=1,value=1234)

# Function to fetch data (Mocking the connection for the script)
@st.cache_data
def get_user_data(uid):
  # I created a simulated result based on my ML code
  return {
      'Frequency':5,
      'Avg Spend': 85.50,
      'Predicted LTV':450.00,
      'Churn Risk':'Low'
  }

  data = get_user_data(user_id)

  # Layout: Key Metrics
  col1,col2,col3,col4 = st.columns(4)
  col1.metric('Orders',data['Frequency'])
  col2.metric('AOV',f'‚Ç¨{data['Avg Spend']}')
  col3.metric('Predicted LTV',f'‚Ç¨{data['Predicted LTV']}',delta='High Value')
  col4.metric('Churn Risk',data['Churn Risk'],delta_color='inverse')

  # Visualizing Behavior (using my HEX #6B46C1)
  st.subheader('User Purchase Trend')
  # Simulated trend data
  trend_data = pd.DataFrame({'Month':['Jan','Feb','Mar'], 'Spend':[40,70,85]})

  fig = px.line(
      trend_data,
      x='Month',
      y='Spend',
      color_discrete_sequence=['#27C7C1'])

  st.plotly_chart(
      fig,
      user_container_width=True
  )
```
